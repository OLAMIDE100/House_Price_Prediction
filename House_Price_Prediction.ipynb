```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_excel(r'C:\Users\OLAMIDE\Desktop\MACHINE_LEARNING_PROJECTS\House_Price_Prediction\Real estate valuation data set.xlsx',header=None)
```

```python
data.columns = data.iloc[0]
```

```python
data = data.drop([0])
data
```

```python
data.info()
```

```python
features = data.drop('Y house price of unit area', axis=1)
features
```

```python
target = data['Y house price of unit area']
```

```python
from sklearn.model_selection import train_test_split
featurestrain, featurestest, targettrain, targettest = train_test_split(features,target,
                                                random_state=1)
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model.fit(featurestrain,targettrain)
predicted = model.predict(featurestest)
```

```python
predicted
```

```python
k = pd.DataFrame({'test':targettest,'predicted':predicted},columns = ['test','predicted'])
k
```

```python
params = pd.Series(model.coef_, index=features.columns)

params
```

```python
from sklearn.utils import resample
np.random.seed(1)
err = np.std([model.fit(*resample(features, target)).coef_
              for i in range(1000)], 0)
print(pd.DataFrame({'effect': params.round(3),
                    'error': err.round(3)}))
```

```python
from sklearn.svm import SVR
mode = SVR(kernel = 'linear',C=100,gamma='auto')
mode.fit(featurestrain,targettrain)
p = mode.predict(featurestest)
```

```python
o = pd.DataFrame({'test':targettest,'predicted':p},columns = ['test','predicted'])
o
```

```python
param = pd.Series(mode.coef_.reshape(6), index=features.columns)

param
```

```python
from sklearn.utils import resample
np.random.seed(1)
err = np.std([mode.fit(*resample(features, target)).coef_.reshape(6)], 0)
print(pd.DataFrame({'effect': param.round(8),
                    'error': err.round(8)}))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(200)
forest.fit(featurestrain,targettrain)
PD = forest.predict(featurestest)
```

```python
I = pd.DataFrame({'test':targettest,'predicted':PD},columns = ['test','predicted'])
I
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
print(mean_absolute_error(targettest,predicted),mean_absolute_error(targettest,p),mean_absolute_error(targettest,PD))
```

```python
print(r2_score(targettest,predicted),r2_score(targettest,p),r2_score(targettest,PD))
```
